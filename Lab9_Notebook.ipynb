```python
import pymysql

def compare_rentals(customer_id):
    connection = pymysql.connect(
        host='MacBook-Air-de-Pablo.local',
        user='root',
        password='MSa75047&',
        database='sakila'
    )

    query_may = f"""
    select count(*) as rental_count_may
    from rentals_may
    where customer_id = {customer_id}
    """

    query_june = f"""
    select count(*) AS rental_count_june
    from rentals_june
    where customer_id = {customer_id}
    """

    try:
        with connection.cursor() as cursor:
            cursor.execute(query_may)
            result_may = cursor.fetchone()
            rentals_may = result_may['rental_count_may']

            cursor.execute(query_june)
            result_june = cursor.fetchone()
            rentals_june = result_june['rental_count_june']

            if rentals_june > rentals_may:
                print(f"Customer {customer_id} borrowed more films in June compared to May.")
            elif rentals_june < rentals_may:
                print(f"Customer {customer_id} borrowed less films in June compared to May.")
            else:
                print(f"Customer {customer_id} borrowed the same number of films in June and May.")

    finally:
        connection.close()
```
